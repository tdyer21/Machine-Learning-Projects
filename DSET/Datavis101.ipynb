# Defining Axes, Ticks, and Grids

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])
ax.grid()

ax.plot(x,y)
```

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(x)
ax2.plot(x,y)
```

# Labels and Annotation

```python
rcParams['figure.figsize'] = 8,4
```

```python
x = range(1,10)
y = [1,2,3,4,.5,4,3,2,1]

plt.bar(x,y)
plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
z = [1,2,3,4,.5]
veh_type = ['bike', 'car', 'skateboard', 'stroller', 'truck']
plt.pie(z, labels=veh_type)
plt.show()
```

```python
mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title("mpg of cars in dataset")

ax.set_xlabel('car names')
ax.set_ylabel('mpg')

ax.annotate('Toyota Corolla', xy=(19,33.9), xytext=(21,35), arrowprops=dict(facecolor='black', shrink=0.05))
```
